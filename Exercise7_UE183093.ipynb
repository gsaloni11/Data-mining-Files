#### Q1:
Find mean, std dev, min , max of unitprice field.

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
UnitPrice = []
with open('OnlineRetail.csv','r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        UnitPrice.append(float(row['UnitPrice']))
        
UnitPrice = np.array(UnitPrice)
print("For the unit price field:")
print("Mean: ", np.mean(UnitPrice))
print("Std.Dev: ", np.std(UnitPrice))
print("Min: ", np.min(UnitPrice))
print("Max: ", np.max(UnitPrice))
```

#### Q2:
Draw histogram of unitprice field using matplotlib

```python
plt.hist(UnitPrice)
plt.title("Unit Price Histogram")
plt.xlabel("Unit Price")
plt.show()
```

#### Q3:
Find mean of unitprice group by country and plot them

```python
mean_unitPrice = {}
with open('OnlineRetail.csv','r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        if(row['Country']) in mean_unitPrice:
            mean_unitPrice[row['Country']].append(float(row['UnitPrice']))
        else:
            mean_unitPrice[row['Country']] = list()
            mean_unitPrice[row['Country']].append(float(row['UnitPrice']))
            
    print("Mean Unit Price for each country\n")
    for i in mean_unitPrice:
        mean_unitPrice[i] = np.mean(np.array(mean_unitPrice[i]))
        print(i," : ",mean_unitPrice[i])
```

```python
fig = plt.figure(figsize=(12,5))
plt.xlabel('Country')
plt.ylabel('Unit Price')
plt.title('Mean Unit Price for each country')
plt.bar(list(mean_unitPrice.keys()), list(mean_unitPrice.values()))
plt.xticks(list(mean_unitPrice.keys()),list(mean_unitPrice.keys()), rotation=90)
plt.show()
```

#### Q4:
Find min and max of unit price group by customerID

```python
customers = {}
with open('OnlineRetail.csv','r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        if (row['CustomerID']) in customers:
            customers[row['CustomerID']].append(float(row['UnitPrice']))
        else:
            customers[row['CustomerID']] = list()
            customers[row['CustomerID']].append(float(row['UnitPrice']))
            
    print("Min and Max values of Unit Price for each customer\n")
    print("CustomerID   MIN        MAX")
    for i in customers:
        customers[i] = np.array(customers[i])
        print(i,"      ",np.min(customers[i]),"     ",np.max(customers[i]))
    
```
