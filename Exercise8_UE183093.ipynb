```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Q1:
Read NIRF data.

```python
df = pd.read_csv('nirf_engg_ranking_india_2020.csv')
df.head()
```

#### Q2:
Group the data statewise and Count the number of colleges in each state.

```python
grouped = df.groupby('state')
print(grouped.size())
```

#### Q3:
Sort the records based on rpc attribute (Research Professional practices).\

```python
df1 = df.sort_values('rpc',ascending=False)
df1.head()
```

#### Q4:
Find the rank of Panjab University as per rpc attribute.

```python
df1['RANK_RPC'] = df1['rpc'].rank(ascending=False)
df1[['name','RANK_RPC']][df1['name']=='Panjab University']
```

#### Q5:
List the colleges whose rpc value is less than Panjab University and rank is higher(rank value is less).

```python
colleges = df[np.logical_and(df['rpc']<df.iloc[71]['rpc'], df['rank']<df.iloc[71]['rank'])]
colleges
```

#### Q6:
Draw the Histogram for colleges based on tlr values by dividing the overall data in 10 bins.

```python
values = []
for lab, row in df.iterrows():
    values.append(row['tlr'])
    
plt.hist(values,bins=10)
plt.title("Colleges based on tlr")
plt.show()
```
